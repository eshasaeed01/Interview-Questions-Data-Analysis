### Question 5:

**For each day of sample collection, determine the proposition of the respondents who were fully satisfied with the performance of the CM. So if there were total 1000 samples on day 1 and 300 out of those said that they are fully satisfied then our answer for that day would be 0.3.**

```python
survey['CM_satisfaction'].unique()
```

```python
CM = survey[survey['CM_satisfaction']=='Fully Satisfied'] # create a new database with existing one
```

```python
a = CM.collection_date.value_counts()
```

```python
b= survey.collection_date.value_counts()
```

```python
c = a/b * 100 # shows the proportion each day
print(c)
```
